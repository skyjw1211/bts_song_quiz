# 정답 맞춰보기

```python
import pandas as pd
import random

df = pd.read_csv('bts_song_data.csv', encoding = 'utf8')

song_ls = df['곡명'].to_list()
d = random.sample(song_ls, 1)
d
```

```python
from kiwipiepy import Kiwi
kiwi = Kiwi()

def lemmatizing(line):
    res = []
    tokens = kiwi.tokenize(line)
    for t in tokens:
        if t.tag[0] not in ['E', 'J'] and t.tag[:2] not in ['SP', 'SC']:
            res.append(t.form)

    return res

def lyrics_tf(lyrics):
    lemma_ls = []
    for line in lyrics:
        line = line.lower()
        lemma_ls.extend(lemmatizing(line))

    lemma_dict = {}
    for lemma in list(set(lemma_ls)):
        lemma_dict[lemma] = lemma_ls.count(lemma)
    
    return lemma_dict
        


```

```python
import pandas as pd
from tqdm import tqdm
# 가사 추출
lyrics_info = []

for song in tqdm(song_ls):
    

    song_name = song
    lyrics = df.loc[df['곡명']==song_name, '가사'].values[0].lower().strip().split('\n')

    song_name_tokens = song_name.lower().split()

    lemma_dict = lyrics_tf(lyrics)

    for line in lyrics:

        line = line.lower()

        # - <제목점수>: 제목과 겹치는 어휘가 있는가, 제목과 유사도?(y/n)
        token_included_cnt = 0
        line_tokens = line.split()
        for t in song_name_tokens:
            if t in line_tokens:
                token_included_cnt += 1
                line_tokens.remove(t)

        title_point = token_included_cnt/len(song_name_tokens)

        # - <길이점수>: 길이가 긴가(y/n) - 어절 수
        len_point = len(line.split())

        # - <후렴점수>: 후렴구인가(중복되는 라인인가)(y/n)
        savi_point = lyrics.count(line)

        # - <max keyword freq>: 전체 가사 중 정보성 높은(잘 사용되지 않는) 어휘가 얼마나 사용되었는가, 최고빈도 어휘의 빈도 수(y/n)
        max_keyword_freq = max(list(map(lambda x: lemma_dict[x], lemmatizing(line))))
        avg_keyword_freq = round(sum(list(map(lambda x: lemma_dict[x], lemmatizing(line))))/len(list(map(lambda x: lemma_dict[x], lemmatizing(line)))),1)
        

        lyrics_info.append((song_name, line, title_point, len_point, savi_point, max_keyword_freq, avg_keyword_freq))

lyric_df = pd.DataFrame(lyrics_info, columns = ['곡명','라인', '제목점수', '길이점수', '후렴점수', '최고빈도점수', '평균빈도점수'])
lyric_df.to_csv('bts_lyrics_data.csv', encoding = 'utf-8-sig', index = 'False')
```
