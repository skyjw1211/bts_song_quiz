# 선수 url 크롤링

```python
from crawler_modules import *
import time
from tqdm import tqdm

base = 'https://www.koreabaseball.com'
```

```python
base_url = 'https://www.koreabaseball.com/Player/Register.aspx'
# print(base_url)
driver, soup = drive(base_url)
```

```python
player_urls = []

for i in range(10):
    driver.find_element_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div/div[1]/ul/li[{i+1}]/a/img').click()
    
    time.sleep(2)

    soup = get_soup(driver, nscroll = 0)

    table = soup.find_all('table', attrs={'class':'tNData'})[2]
    for player in table.find_all('a'):
        player_urls.append(player.get('href'))
    
    table = soup.find_all('table', attrs={'class':'tNData'})[3]
    for player in table.find_all('a'):
        player_urls.append(player.get('href'))
    
    table = soup.find_all('table', attrs={'class':'tNData'})[4]
    for player in table.find_all('a'):
        player_urls.append(player.get('href'))
    
    table = soup.find_all('table', attrs={'class':'tNData'})[5]
    for player in table.find_all('a'):
        player_urls.append(player.get('href'))

print(len(player_urls))
player_urls
```

```python
print(len(set(player_urls)))
```

```python
player_infos = []
import re

for player_url in tqdm(player_urls):
    driver.get(base+player_url)
    soup = get_soup(driver, nscroll = 0)
    time.sleep(1)
    text = soup.find_all('div', attrs={'class':'player_basic'})[0].text.strip()

    name = text.split("선수명: ")[1].split("\n")[0]
    number = text.split("등번호: ")[1].split("\n")[0]
    birthday = text.split("생년월일: ")[1].split("\n")[0]
    position = text.split("포지션: ")[1].split("\n")[0]
    height_weight = text.split("신장/체중: ")[1].split("\n")[0]
    career = text.split("경력: ")[1].split("\n")[0]
    signing_fee = text.split("입단 계약금: ")[1].split("\n")[0]
    salary = text.split("연봉: ")[1].split("\n")[0]
    draft = text.split("지명순위: ")[1].split("\n")[0]
    year_of_entry = text.split("입단년도: ")[1].split("\n")[0]

    number = number.lstrip('No.')
    height = height_weight.split('/')[0]
    weight = height_weight.split('/')[1]
    position, throwing_hand = re.match('(.*)\((.*)\)', position).groups()
    



    result = (name, number, birthday, position, height_weight, career, signing_fee, salary, draft, year_of_entry)
    player_infos.append(result)

player_infos
```

```python
# 28명씩
import re
teams = ['SSG', '키움', 'LG', 'KT', 'KIA', 'NC', '삼성', '롯데', '두산', '한화']

player_infos2 = []

i = 0

for name, number, birthday, position, height_weight, career, signing_fee, salary, draft, year_of_entry in player_infos:
    team = teams[i//28]
    number = number.lstrip('No.')
    height = height_weight.split('/')[0]
    weight = height_weight.split('/')[1]
    position, throwing_hand = re.match('(.*)\((.*)\)', position).groups()

    player_infos2.append((name, team, number, birthday, position, throwing_hand, height, weight, career, signing_fee, salary, draft, year_of_entry))

    i += 1

player_infos2
```

```python
df = pd.DataFrame(player_infos2, columns= ('이름', '팀', '번호', '생일', '포지션', '투타유형', '키', '몸무게', '경력', '입단 계약금', '연봉', '지명 순위', '입단년도'))
df.to_csv('baseball_player.csv', encoding = 'utf-8-sig', index = 'False')
```
